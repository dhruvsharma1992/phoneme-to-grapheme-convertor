```python
from difflib import SequenceMatcher

def getHighestMatch(masterList,word):
    prev_diff_val = 0
    match_word = ''
    for x in masterList:
        new_diff_val = similar(str(word),x)
        if new_diff_val > prev_diff_val:
            match_word = x
            prev_diff_val = new_diff_val
    return prev_diff_val,match_word

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

def getPossibleGraphemes(arpabets, arpabetDict):
    temp_list = []
    for arp in arpabets:
        if arp in arpabetDict:
            temp_list.append(arpabetDict[arp])
    return temp_list
match_list = []
```

```python
def addCur(grp,x,templst,masterList):
#     print(x)
#     print(y)
#     print(templst)
#     print(masterList)
#     print("^"*10)
    if x == len(grp):
        masterList.append(templst)
        return
    for i in grp[x]:
        templ2 = list(templst)
        templ2.append(i)
        addCur(grp,x+1,templ2,masterList)
    del templst
    del templ2
    return
```
